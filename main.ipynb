##Imports Libraries:

- Gradio: For creating the web interface where users can upload images and view results.
- PIL (Pillow): For handling image processing tasks like opening and saving images.
- Subprocess: To execute shell commands from within Python.
- OS: For interacting with the operating system, such as checking if files exist.

## Define Paths:

- Input Image Path: The location where the uploaded image will be saved temporarily.
- Output Image Path: The location where the processed image (result of YOLO detection) will be saved.

## Process Image Function:

- Save the Image: Saves the uploaded image to the specified path.
Run YOLO Detection: Executes the YOLO command to perform object detection on the saved image. It captures any output or errors for debugging.
- Handle Errors: If YOLO fails, it provides a red placeholder image to indicate an error.
- Check Output File: Verifies if the output image from YOLO exists.
- Open and Return Image: If the output file exists and can be opened, it returns the processed image. If not, it returns a placeholder image (yellow if the file cannot be opened, blue if the file is not found).

## Create Gradio Interface:

- Setup Interface: Configures Gradio to use the process_image function. - Specifies that users will upload images and receive processed images in return.
- Launch the App: Starts the Gradio app, making it accessible to users through a web interface.

```python
import gradio as gr
from PIL import Image as PILImage
import subprocess
import os
import shutil

# Define paths
runs_base_path = '/content/runs'
detect_base_path = os.path.join(runs_base_path, 'detect')
input_image_path = '/content/uploaded_image.jpg'

def ensure_directories_exist():
    # Check if the runs folder exists, if not, create it
    if not os.path.exists(runs_base_path):
        os.makedirs(detect_base_path)
    # Check if the detect folder exists, if not, create it
    elif not os.path.exists(detect_base_path):
        os.makedirs(detect_base_path)

def get_new_prediction_folder():
    # Ensure base folders exist before proceeding
    ensure_directories_exist()

    # Check how many prediction folders currently exist
    existing_folders = [folder for folder in os.listdir(detect_base_path) if folder.startswith('predict')]
    next_folder_number = len(existing_folders) + 1
    new_folder_name = f'predict{next_folder_number}' if next_folder_number > 1 else 'predict'
    return os.path.join(detect_base_path, new_folder_name)

def process_image(image):
    # Save the uploaded image to a file
    image.save(input_image_path)

    # Determine the new prediction folder
    new_prediction_folder = get_new_prediction_folder()
    output_image_path = os.path.join(new_prediction_folder, 'uploaded_image.jpg')

    # Run YOLO detection
    try:
        result = subprocess.run([
            'yolo',
            'task=detect',
            'mode=predict',
            f'model=/content/best.pt',
            'conf=0.25',
            f'source={input_image_path}',
            f'project={detect_base_path}',
            f'name={os.path.basename(new_prediction_folder)}',
            'exist_ok=True'
        ], capture_output=True, text=True, check=True)
        print("YOLO command output:", result.stdout)
        print("YOLO command error (if any):", result.stderr)
    except subprocess.CalledProcessError as e:
        print(f"Error running YOLO: {e}")
        return PILImage.new("RGB", (100, 100), color="red")

    # Check if the output file exists
    if os.path.exists(output_image_path):
        try:
            new_image = PILImage.open(output_image_path)
            return new_image
        except IOError:
            print("Error opening output image file.")
            return PILImage.new("RGB", (100, 100), color="yellow")
    else:
        print(f"Output image file not found at {output_image_path}.")
        return PILImage.new("RGB", (100, 100), color="blue")

# Create a Gradio interface
iface = gr.Interface(
    fn=process_image,
    inputs=gr.Image(type="pil"),
    outputs=gr.Image(type="pil"),
    title="Helmet and Number Plate Detection",
    description="Upload an image to perform detection using YOLO."
)

iface.launch()
```

- The script imports necessary libraries like Gradio, PIL (for image handling), subprocess (for running YOLO commands), and other utilities.
- The input_image_path is defined as the location where the uploaded image will be saved.
- **ensure_directories_exist Function:** This function checks if the runs and detect folders exist. If they don't, it creates them.
This ensures that your script won't fail if these directories are missing.
- **Modified get_new_prediction_folder Function:** Before determining the next prediction folder, it ensures the base directories exist by calling ensure_directories_exist.
- The detect_base_path is defined as the directory where YOLO outputs will be stored.
- The get_new_prediction_folder function identifies how many prediction folders currently exist and determines the name of the next folder.
- Inside this function, it checks the runs/detect/ directory for existing folders named "predict", "predict2", etc.
- It calculates the next folder name (e.g., predict3) based on the number of existing folders and returns the path for the new folder.
- The process_image function starts by saving the uploaded image to a file at the specified path.
- It then determines the new prediction folder by calling get_new_prediction_folder.
- The path for the predicted output image is set, including the folder created above.
- The script runs a YOLO detection command using subprocess.run, specifying the model and source image, along with output paths.
- It captures and prints the output and errors (if any) from the YOLO command.
- If there’s an error in running YOLO, it catches the exception and returns a red placeholder image.
- It checks whether the output image exists at the expected path after YOLO runs.
- If the image exists, it attempts to load it using PIL and returns the image.
- If there's an error in loading the image, a yellow placeholder image is returned.
- If the output image isn’t found, a blue placeholder image is returned.
- A Gradio interface is created, specifying the image input, image output, and relevant descriptions.
- The Gradio interface is launched to allow users to interact with the model by uploading images.

# For Image

```python
from fastapi import FastAPI, File, UploadFile
from fastapi.responses import FileResponse
from PIL import Image as PILImage
import subprocess
import os
import shutil
from io import BytesIO
from pyngrok import ngrok
import nest_asyncio
import uvicorn

app = FastAPI()

# Define paths
runs_base_path = '/content/runs'
detect_base_path = os.path.join(runs_base_path, 'detect')
input_image_path = '/content/uploaded_image.jpg'

def ensure_directories_exist():
    # Check if the runs folder exists, if not, create it
    if not os.path.exists(runs_base_path):
        os.makedirs(detect_base_path)
    # Check if the detect folder exists, if not, create it
    elif not os.path.exists(detect_base_path):
        os.makedirs(detect_base_path)

def get_new_prediction_folder():
    # Ensure base folders exist before proceeding
    ensure_directories_exist()

    # Check how many prediction folders currently exist
    existing_folders = [folder for folder in os.listdir(detect_base_path) if folder.startswith('predict')]
    next_folder_number = len(existing_folders) + 1
    new_folder_name = f'predict{next_folder_number}' if next_folder_number > 1 else 'predict'
    return os.path.join(detect_base_path, new_folder_name)

def process_image(image: PILImage.Image):
    # Save the uploaded image to a file
    image.save(input_image_path)

    # Determine the new prediction folder
    new_prediction_folder = get_new_prediction_folder()
    output_image_path = os.path.join(new_prediction_folder, 'uploaded_image.jpg')

    # Run YOLO detection
    try:
        result = subprocess.run([
            'yolo',
            'task=detect',
            'mode=predict',
            f'model=/content/best.pt',
            'conf=0.25',
            f'source={input_image_path}',
            f'project={detect_base_path}',
            f'name={os.path.basename(new_prediction_folder)}',
            'exist_ok=True'
        ], capture_output=True, text=True, check=True)
        print("YOLO command output:", result.stdout)
        print("YOLO command error (if any):", result.stderr)
    except subprocess.CalledProcessError as e:
        print(f"Error running YOLO: {e}")
        return None

    # Check if the output file exists
    if os.path.exists(output_image_path):
        return output_image_path
    else:
        print(f"Output image file not found at {output_image_path}.")
        return None

@app.post("/detect/")
async def detect_helmet_license_plate(file: UploadFile = File(...)):
    # Read the image uploaded by the user
    image_bytes = await file.read()
    image = PILImage.open(BytesIO(image_bytes))

    # Process the image using YOLO
    output_image_path = process_image(image)

    if output_image_path:
        # Return the processed image
        return FileResponse(output_image_path, media_type="image/jpeg")
    else:
        # In case of error, return a 500 response with a placeholder image
        error_image = PILImage.new("RGB", (100, 100), color="red")
        buf = BytesIO()
        error_image.save(buf, format="JPEG")
        buf.seek(0)
        return FileResponse(buf, media_type="image/jpeg")

# Setup ngrok to expose the local server
ngrok_tunnel = ngrok.connect(8000)
print('Public URL:', ngrok_tunnel.public_url)

# Allow running uvicorn in a Jupyter environment
nest_asyncio.apply()

# Start the server
uvicorn.run(app, port=8000)
```

# For Video

```python
from fastapi import FastAPI, File, UploadFile
from fastapi.responses import FileResponse
import os
import subprocess
from pyngrok import ngrok
import nest_asyncio
import uvicorn

app = FastAPI()

# Define paths
runs_base_path = '/content/runs'
detect_base_path = os.path.join(runs_base_path, 'detect')
input_video_path = '/content/uploaded_video.mp4'
intermediate_video_path = '/content/intermediate_video.avi'  # Assuming YOLO might output in AVI format
output_video_path = '/content/output_video.mp4'  # Final output video in MP4 format

def ensure_directories_exist():
    # Ensure base directories exist
    if not os.path.exists(runs_base_path):
        os.makedirs(detect_base_path)
    elif not os.path.exists(detect_base_path):
        os.makedirs(detect_base_path)

def get_new_prediction_folder():
    # Ensure base folders exist before proceeding
    ensure_directories_exist()

    # Check how many prediction folders currently exist
    existing_folders = [folder for folder in os.listdir(detect_base_path) if folder.startswith('predict')]
    next_folder_number = len(existing_folders) + 1
    new_folder_name = f'predict{next_folder_number}' if next_folder_number > 1 else 'predict'
    return os.path.join(detect_base_path, new_folder_name)

def convert_to_mp4(input_path: str, output_path: str):
    """Convert video to MP4 using ffmpeg."""
    try:
        result = subprocess.run([
            'ffmpeg', '-i', input_path, '-c:v', 'libx264', '-c:a', 'aac', '-strict', 'experimental', output_path
        ], capture_output=True, text=True, check=True)
        print("FFmpeg output:", result.stdout)
        print("FFmpeg error (if any):", result.stderr)
    except subprocess.CalledProcessError as e:
        print(f"Error converting video to MP4: {e}")
        raise

def process_video(video_path: str):
    # Determine the new prediction folder
    new_prediction_folder = get_new_prediction_folder()

    # YOLO video detection command
    try:
        result = subprocess.run([
            'yolo',
            'task=detect',
            'mode=predict',
            f'model=/content/best.pt',
            'conf=0.25',
            f'source={video_path}',
            f'project={detect_base_path}',
            f'name={os.path.basename(new_prediction_folder)}',
            'exist_ok=True'
        ], capture_output=True, text=True, check=True)
        print("YOLO command output:", result.stdout)
        print("YOLO command error (if any):", result.stderr)
    except subprocess.CalledProcessError as e:
        print(f"Error running YOLO: {e}")
        return None

    # Check the folder contents to see if a video was created
    folder_contents = os.listdir(new_prediction_folder)
    print(f"Contents of the prediction folder: {folder_contents}")

    # YOLO might save the video in various formats, e.g., AVI
    predicted_video_path = os.path.join(new_prediction_folder, 'result.avi')  # Adjust filename and extension if needed
    print(f"Predicted video path: {predicted_video_path}")

    # Check if the predicted video exists
    if os.path.exists(predicted_video_path):
        print(f"Predicted video found at {predicted_video_path}")
        # Convert to MP4
        convert_to_mp4(predicted_video_path, output_video_path)
        return output_video_path
    else:
        print(f"Processed video not found at {predicted_video_path}.")
        return None

@app.post("/detect-video/")
async def detect_helmet_license_plate_video(file: UploadFile = File(...)):
    # Save the uploaded video
    with open(input_video_path, "wb") as f:
        f.write(await file.read())
    print(f"Uploaded video saved at {input_video_path}")

    # Process the video using YOLO and convert to MP4
    processed_video_path = process_video(input_video_path)

    if processed_video_path:
        # Return the processed video file in MP4 format
        print(f"Returning processed video: {processed_video_path}")
        return FileResponse(processed_video_path, media_type="video/mp4")
    else:
        # In case of error, return a 500 response
        return {"error": "Video processing failed"}

# Setup ngrok to expose the local server
ngrok_tunnel = ngrok.connect(8000)
print('Public URL:', ngrok_tunnel.public_url)

# Allow running uvicorn in a Jupyter environment
nest_asyncio.apply()

# Start the server
uvicorn.run(app, port=8000)
```
